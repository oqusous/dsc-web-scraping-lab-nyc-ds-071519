## Retrieving Titles

To start, write a function that extracts the titles of the books on a given page. The input for the function should be the `soup` for the HTML of the page.

````python
from bs4 import BeautifulSoup
import requests
````

````python
html_page = requests.get('http://books.toscrape.com/') #Make a get request to retrieve the page
soup = BeautifulSoup(html_page.content, 'html.parser') #Pass the page contents to beautiful soup for parsing
````

````python
soup.prettify
````

````python
title_filter1 = soup.find('h3')
title_filter1
````

````python
title_filter2 = title_filter1.find('a')['title']
title_filter2
````

````python
title_access = soup.find('h3').find('a').attrs['title']
title_access
````

````python
def retrieve_titles(soup):
    titles = []
    for title in soup.findAll('h3'):
        titles.append(title.find('a')['title'])
    return titles
    #Your code here
````

````python
retrieve_titles(soup)
````

````python
def retrieve_ratings(soup):
    rating_filter2 = (soup.findAll('p', 'star-rating'))
    lists = []
    for i in rating_filter2:
        string = str(i)
        lists.append(string[22:-137])
    star_dict = {'One': 1, 'Two': 2, 'Three':3, 'Four': 4, 'Five':5} #Manually create a dictionary to translate to numeric
    star_ratings = [star_dict[s] for s in lists]
    return star_ratings
````

## Retrieve Prices

Now write a function to retrieve the prices on a given page. The function should take in the `soup` from the given page and return a list of prices formatted as floats.

````python
filter_prices = soup.findAll('p', class_="price_color")
filter_prices
````

````python
def retrieve_prices(soup):
    prices = []
    prices_cleaned = []
    for p in soup.findAll('p', class_="price_color"):
        prices.append(p.text) 
    for p in prices:
        prices_cleaned.append(float(p[1:]))
    return prices_cleaned
    #Keep cleaning it up
    #Your code here
````

````python
retrieve_prices(soup)
````

## Retrieve Availability

Write a function to retrieve whether each book is available or not. The function should take in the `soup` from a given html page and return a list of the availability for each book.

````python
def retrieve_availabilities(soup):
    avails = []
    for a in soup.findAll('p', class_="instock availability"):
        b = a.text.strip()
        avails.append(b)#Finalize the selection
    return avails
    #Your code here
````

````python
retrieve_availabilities(soup)
````

## Create a Script to Retrieve All the Books From All 50 Pages

Finally, write a script to retrieve all of the information from all 50 pages of the books.toscrape.com website. 

````python
import pandas as pd
````

````python
#Your code here
titles_all = []
star_ratings_all = []
prices_all = []
avails_all = []

for i in range(1,51):
    if i == 1:
        url = 'http://books.toscrape.com/'
    else:
            url = "http://books.toscrape.com/catalogue/page-{}.html".format(i)
    html_page_all = requests.get(url)
    soup_all = BeautifulSoup(html_page_all.content, 'html.parser')
    titles_all += retrieve_titles(soup_all)
    star_ratings_all += retrieve_ratings(soup_all)
    prices_all += retrieve_prices(soup_all)
    avails_all += retrieve_availabilities(soup_all)
df = pd.DataFrame([titles_all, star_ratings_all, prices_all, avails_all]).transpose()
df.columns = ['Title', 'Star_Rating', 'Price_(pounds)', 'Availability']
print(len(df))
df
````
